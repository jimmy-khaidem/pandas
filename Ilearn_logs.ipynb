```python
import pandas as pd
ilearnonlinelogs=pd.read_excel('S:\\Aroha_tech\\assignmentQ\\powerbi\\Ilearnonlinelogs_ETL.xlsx',sheet_name='Ilearnonlinelogs')
ilearnonlinelogsdf=pd.DataFrame(ilearnonlinelogs)
ilearnonlinelogsdf
```

```python
filtereddf=ilearnonlinelogsdf[ilearnonlinelogsdf['path']!='/']
filtereddf=filtereddf.reset_index(drop=True)
filtereddf
```

```python
rejected=ilearnonlinelogsdf[ilearnonlinelogsdf['path']=='/']
rejected
```

```python
withoutchapterlst=[]
withchapterlst=[]
for (columnName, columnData) in filtereddf.iteritems():
    if columnName=='path':
        for i in columnData.values:
            if i.count('/')<=3:
                withoutchapterlst.append(i)
            else:
                withchapterlst.append(i)
print(withchapterlst)
```

```python
withindex=[]
for i in withchapterlst:
    a=filtereddf[filtereddf['path']==i].index.values
    s=list(a)
    withindex.append(s) 
withindex
```

```python
storelst=[]
for i in withindex:
    if type(i)==int:
        storelst.append(i)
    if type(i)==list:
        for j in i:
            storelst.append(j)
storelst=list(set(storelst))
```

```python
storelst=[int(i) for i in storelst]
course_log_unwanted=filtereddf.drop(filtereddf.index[storelst],axis=0)
course_log_unwanted
```

```python
cleanind=[]
for i in withoutchapterlst:
    b=filtereddf[filtereddf['path']==i].index.values
    v=list(b) 
    cleanind.append(v)
cleanind
```

```python
savelst=[]
for i in cleanind:
    if type(i)==int:
        savelst.append(i)
    if type(i)==list:
        for j in i:
            savelst.append(j)
savelst=list(set(savelst))
```

```python
savelst=[int(i) for i in savelst]
course_log=filtereddf.drop(filtereddf.index[savelst],axis=0)
course_log
```
